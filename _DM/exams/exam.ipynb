```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AirQualityUCI.csv", delimiter=";")
print("Shapes:", df.shape)
df.head(10)
print("Nullrows:", df.shape[0] - df.dropna().shape[0])
df.info()
df.describe()
```

```python
df = (
    df.dropna(how="all", axis=0)
    .dropna(how="all", axis=1)
    .dropna(thresh=df.shape[0] / 3, axis=1)
    .apply(lambda x: x.fillna(x.mean()) if x.dtype == float else x)
)
print("Remaining any Null rows:", df.shape[0] - df.dropna().shape[0])
df
```

```python
df = df.drop("Time", axis=1)
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
df = df.groupby("Date").mean()
display(df)
```

```python
from sklearn.preprocessing import KBinsDiscretizer

bins = 2
cols = df.columns
df = KBinsDiscretizer(
    bins, encode="onehot-dense", strategy="kmeans", subsample=None
).fit_transform(df)

display(df)
df = df > 0
display(df)
```

```python
new_names = []
for col in cols:
    new_names.append(col + "_low")
    new_names.append(col + "_high")

df = pd.DataFrame(df, columns=new_names)
display(df)
```
