```python
# drop the index column because it's not useful
df.drop("Index", axis=1, inplace=True)
```

```python
# One Hot Encoding of "Sex" column
df = pd.get_dummies(df, columns=["Sex"])
```

```python
sns.heatmap(df.corr().abs(), cmap="Blues", annot=True)
```

```python
# Drop "Length" because is very highly correlated with "Diameter" but is less correlated with "Rings"
# Drop "Whole weight" because is highy correlated with the other weights

df = df.drop(["Length", "Whole weight"], axis=1)
df = df.dropna()
print(df.shape[0])
df.head(5)
```


4. find the best classification scheme considering two classification methoods, find the best hyperparameters using cross validation; the optimization must be focused on the f1_macro measure . . . . . . . . . . . . . . . . . . .4pt


```python
from sklearn.model_selection import train_test_split

X = df.drop("Rings", axis=1)
y = df["Rings"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline

# iterate over multiple estimators using a Pipeline
param_grid = {
    "criterion": ["gini", "entropy", "log_loss"],
    "max_depth": [10, 30, 40, 50, 70, 100, None],
}

grid_rf = GridSearchCV(
    estimator=RandomForestClassifier(n_jobs=-1),
    param_grid=param_grid,
    scoring="f1_macro",
    cv=5,
    n_jobs=-1,
)
grid_rf.fit(X_train, y_train)

print(
    "Params: {}\nScore: {}\n".format(
        grid_rf.best_params_,
        grid_rf.score(X_test, y_test),
    )
)
```

```python
param_grid = {
    "weights": ["uniform", "distance"],
    "metric": [
        "cityblock",
        "cosine",
        "euclidean",
        "l1",
        "l2",
        "manhattan",
        "nan_euclidean",
    ],
    "n_neighbors": [*range(1, 30)],
}


grid_knn = GridSearchCV(
    estimator=KNeighborsClassifier(n_jobs=-1),
    param_grid=param_grid,
    scoring="f1_macro",
    cv=5,
    n_jobs=-1,
)
grid_knn.fit(X_train, y_train)

print(
    "Params: {}\nScore: {}\n".format(
        grid_knn.best_params_,
        grid_knn.score(X_test, y_test),
    )
)
```


5. Show the performance measures and the confusion matrices for the best
hyperparameters of each model . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . 2pt


```python
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

y_pred_rf = grid_rf.best_estimator_.predict(X_test)
print("RandomForestClassifier:\n", classification_report(y_test, y_pred_rf))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, cmap="Blues")
plt.show()
```

```python
y_pred_knn = grid_knn.best_estimator_.predict(X_test)
print("KNeighborsClassifier:\n", classification_report(y_test, y_pred_knn))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_knn, cmap="Blues")
plt.show()
```
